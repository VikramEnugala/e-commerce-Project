```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
# Create an empty list to store all mobile Names:
Mobile_Brand = []

# Loop through pages 1 to 21
for page in range(1, 22):

     # URL of the Flipkart search page
    url = f"https://www.flipkart.com/search?q=mobiles+under+50000&page={page}"
    driver.get(url)   
    Mobile_Brands = driver.find_elements(By.CSS_SELECTOR,".RG5Slk")
    for mobile in Mobile_Brands:
        Mobile_Brand.append(mobile.text.strip())
        
    

print(len(Mobile_Brand))
```

```python
# Create an empty list to store all mobile prices
Mobile_Prices = []

# Loop through pages 1 to 21
for page in range(1, 23):

    # URL of the Flipkart search page
    url = f"https://www.flipkart.com/search?q=mobiles+under+50000&page={page}"

    # Open the webpage
    driver.get(url)

    # Find all mobile prices
    prices = driver.find_elements(By.CSS_SELECTOR, ".hZ3P6w.DeU9vF")

    # Store the mobile prices
    for price in prices:
        Mobile_Prices.append(price.text.strip())

# Print the total number of prices scraped
print(len(Mobile_Prices))
```

```python
# Create an empty list to store all mobile ratings
Ratings = []

# Loop through pages 1 to 21
for page in range(1, 22):

    # URL of the Flipkart search page
    url = f"https://www.flipkart.com/search?q=mobiles+under+50000&page={page}"

    # Open the webpage
    driver.get(url)

    # Find all mobile ratings
    mobile_ratings = driver.find_elements(By.CSS_SELECTOR, ".CjyrHS")

    # Store the mobile ratings
    for rating in mobile_ratings:
        Ratings.append(rating.text.strip())

# Print the total number of ratings scraped
print(len(Ratings))
```

```python
# Create an empty list to store all mobile reviews
Reviews = []

# Loop through pages 1 to 21
for page in range(1, 22):

    # URL of the Flipkart search page
    url = f"https://www.flipkart.com/search?q=mobiles+under+50000&page={page}"

    # Open the webpage
    driver.get(url)

    # Find all mobile reviews
    mobile_reviews = driver.find_elements(By.CSS_SELECTOR, ".PvbNMB")

    # Store the mobile reviews
    for review in mobile_reviews:
        Reviews.append(review.text.strip())

# Print the total number of reviews scraped
print(len(Reviews))
```

```python
# Create an empty list to store all mobile descriptions
Descriptions = []

# Loop through pages 1 to 21
for page in range(1, 22):

    # URL of the Flipkart search page
    url = f"https://www.flipkart.com/search?q=mobiles+under+50000&page={page}"

    # Open the webpage
    driver.get(url)

    # Find all mobile descriptions
    mobile_descriptions = driver.find_elements(By.CSS_SELECTOR, ".CMXw7N")

    # Store the mobile descriptions
    for description in mobile_descriptions:
        Descriptions.append(description.text.strip())

# Print the total number of descriptions scraped
print(len(Descriptions))
```

```python
print("Mobile_Brand:", len(Mobile_Brand))
print("Mobile_Prices:", len(Mobile_Prices))
print("Ratings:", len(Ratings))
print("Reviews:", len(Reviews))
print("Descriptions:", len(Descriptions))
```

```python
# Find the maximum length
max_length = max(len(Mobile_Brand), len(Mobile_Prices), len(Ratings), len(Reviews), len(Descriptions))

# Add None to Mobile_Brand until it reaches max_length
while len(Mobile_Brand) < max_length:
    Mobile_Brand.append(None)

# Add None to Mobile_Prices until it reaches max_length
while len(Mobile_Prices) < max_length:
    Mobile_Prices.append(None)

# Add None to Ratings until it reaches max_length
while len(Ratings) < max_length:
    Ratings.append(None)

# Add None to Reviews until it reaches max_length
while len(Reviews) < max_length:
    Reviews.append(None)

# Add None to Descriptions until it reaches max_length
while len(Descriptions) < max_length:
    Descriptions.append(None)
```

```python
#Filling Null Values By Using Mode Method
df["Mobile_Prices"] = df["Mobile_Prices"].fillna(df["Mobile_Prices"].mode()[0])
df["Ratings"] = df["Ratings"].fillna(df["Ratings"].mode()[0])
df["Reviews"] = df["Reviews"].fillna(df["Reviews"].mode()[0])
df["Mobile_Brand"] = df["Mobile_Brand"].fillna(df["Mobile_Brand"].mode()[0])
df["Descriptions"] = df["Descriptions"].fillna(df["Descriptions"].mode()[0])
```

```python
#Converting Categorical datatype to Numerical Datatype
df["Mobile_Prices"] = (
    df["Mobile_Prices"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Mobile_Prices"] = pd.to_numeric(df["Mobile_Prices"], errors="coerce")

df["Ratings"] = pd.to_numeric(df["Ratings"], errors="coerce")
```

```python
df["Reviews"] = (df["Reviews"].astype(str).str.extract(r"(\d+)")[0])

df["Reviews"] = pd.to_numeric(df["Reviews"], errors="coerce")
df["Reviews"]
```
